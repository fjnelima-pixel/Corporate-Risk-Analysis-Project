```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Year": [2021, 2022, 2023],
    "Revenue": [120_000_000, 135_000_000, 150_000_000],
    "EBITDA": [22_000_000, 24_500_000, 25_000_000],
    "Net_Income": [8_000_000, 9_500_000, 9_000_000],
    "Total_Debt": [40_000_000, 42_000_000, 45_000_000],
    "Cash": [6_000_000, 5_000_000, 4_500_000],
    "Current_Assets": [35_000_000, 38_000_000, 40_000_000],
    "Current_Liabilities": [22_000_000, 24_000_000, 26_500_000],
}
df = pd.DataFrame(data)
df
```

```python
df["EBITDA_Margin"] = df["EBITDA"] / df["Revenue"]
df["Debt_to_EBITDA"] = df["Total_Debt"] / df["EBITDA"]
df["Interest_Coverage"] = df["EBITDA"] / 5_000_000   # assume 5m interest expense
df["Current_Ratio"] = df["Current_Assets"] / df["Current_Liabilities"]
df["Net_Debt"] = df["Total_Debt"] - df["Cash"]
df["Net_Debt_to_EBITDA"] = df["Net_Debt"] / df["EBITDA"]

df.round(3)
```
